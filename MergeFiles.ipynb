```python
import pandas as pd
import numpy as np
```

```python
headers = ['var1','var2','var3','var4','var5','var6','o1','o2']
time_header = ['Total Time(s)','Time_o1(s)','Time_o2(s)']
sorted_headers = time_header + headers
```

```python
base_folder = 'final/Metamodels/'
```

```python
def read_sample_file(file, parent=base_folder, header=headers, sep=','):
    df = pd.read_csv(parent + file, sep=sep, header=None, names=header, nrows=100)
    for h in time_header:
        df[h] = pd.DataFrame(np.zeros((100, 1)))
    
    
    return df[sorted_headers]
```

```python
sample_files = ['truss_sample1.csv', 'truss_sample2.csv', 'truss_sample3.csv']
sample_dfs = [read_sample_file(file) for file in sample_files]
```

```python
sample_dfs[0].head()
```

```python
metamodels = ['GPR', 'RF', 'MLP']
models = ['NSGAII', 'Random', 'SMPSO']
nruns = 3
```

```python
def metamodels_files(metamodels=metamodels, models=models, nruns=nruns, parent=base_folder):
    return [parent + metamodel + '_' + model + "_results_0" + str(i) + ".csv"
                for i in range(1, nruns+1) 
                    for model in models 
                        for metamodel in metamodels ]
```

```python
metamodels_filepath = metamodels_files()
metamodels_filepath
```

```python
metamodels_dfs = [pd.read_csv(f, sep=',') for f in metamodels_filepath]
```

```python
def join(data, sample): 
    result = pd.concat([sample, data], axis=0)
    result.reset_index(inplace=True)
    return result
```

```python
new_metamodels_paths = metamodels_files(parent='final/Analises/')
new_metamodels_paths
```

```python
new_metamodels_dfs = []
for i_f, f in enumerate(metamodels_filepath):
    if ("results_01.csv") in f:
        new_metamodels_dfs += [join(metamodels_dfs[i_f], sample_dfs[0])]
    elif ("results_02.csv") in f:
        new_metamodels_dfs += [join(metamodels_dfs[i_f], sample_dfs[1])]
    elif ("results_03.csv") in f:
        new_metamodels_dfs += [join(metamodels_dfs[i_f], sample_dfs[2])]
    else:
        print("WROOOONG")
```

```python
len(new_metamodels_dfs)
```

```python
for i in map(lambda x: len(x), new_metamodels_dfs):
    if i < 225:
        print("Unexpected size!!")
    
```

```python
for i in range(len(new_metamodels_dfs)):
    new_metamodels_dfs[i].to_csv(new_metamodels_paths[i])
```
